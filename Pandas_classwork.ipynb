##### Handling Null
+ In Python null valuses showed with None
+ In Numpy it showed with np.NaN
+ In Pandas pd.Na

```python
import numpy as np
```

```python
dnull = [ [1,2,3,'M',7],
         [np.nan,5,6,np.nan,7],
         [8,np.nan,9,'F',10],
         [np.nan,np.nan,np.nan,'M',11] 
        ]

dframe = pd.DataFrame(dnull, columns=['a','b','c','d','e'])
dframe
```

```python
dframe.info()
```

```python
dframe.isnull().sum()
```

```python
dframe.dropna()
```

```python
dframe.dropna(axis = 1)
```

###### Handling Null values with the mean and mode
+ We can use mean/avg of table to fill Null values
+ And for the object column we can use Mode of column to fill Null values
+ smtimes we can have multiple mode value so they are assigned with the 0,1,2 inside the []

```python
avg_a = dframe['a'].mean()
avg_b = dframe['b'].mean()
avg_c = dframe['c'].mean()
avg_a,avg_b,avg_c
```

```python
dframe['d'].mode()
```

```python
mode_d = dframe['d'].mode()[0]
mode_d
```

```python
dframe.fillna({
    'a' : avg_a,
    'b' : avg_b,
    'c' : avg_c,
    'd' : mode_d
} )
```

```python
dframe
```
